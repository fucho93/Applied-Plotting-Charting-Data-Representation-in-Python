```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib notebook
```

```python
temp = pd.read_csv('C:\\Users\\rafae\\Downloads\\temp_91_16.csv')
rain = pd.read_csv('C:\\Users\\rafae\\Downloads\\rain_91_16.csv')
livestock = pd.read_excel('C:\\Users\\rafae\\Downloads\\pecuario.xls',skip_footer = 5)
```

```python
temp = temp.iloc[:,:3]
rain = rain.iloc[:,:3]
rain.columns = ["Rainfall (mm)", "Year","Month"]
temp.columns = ["Temperature (°C)","Year","Month"]
rain["Year"] = rain["Year"].astype(int)
temp["Year"] = temp["Year"].astype(int)
rain["Month"] = rain["Month"].str.strip()
temp["Month"] = temp["Month"].str.strip()
```

```python
min_rain_y = rain.groupby("Year").min()
max_rain_y = rain.groupby("Year").max()
mean_rain_y = rain.groupby("Year").mean()
min_temp_y = temp.groupby("Year").min()
max_temp_y = temp.groupby("Year").max()
mean_temp_y = temp.groupby("Year").mean()
```

```python
merge_r_y = pd.merge(min_rain_y,max_rain_y,left_index = True, right_index = True)
merge_r_y = pd.merge(merge_r_y,mean_rain_y,left_index = True, right_index = True)
merge_r_y.columns = ["Min","Month_Min","Max","Month_Max","Mean"]

merge_t_y = pd.merge(min_temp_y,max_temp_y,left_index = True, right_index = True)
merge_t_y = pd.merge(merge_t_y,mean_temp_y,left_index = True, right_index = True)
merge_t_y.columns = ["Min","Month_Min","Max","Month_Max","Mean"]
```

```python
value_25 = rain_y.describe().loc["25%",:].values[0]
value_75 = rain_y.describe().loc["75%",:].values[0]
```

```python
rain_b_25 =rain_y[rain_y['Rainfall (mm)'] <= value_25]
rain_a_75 =rain_y[rain_y['Rainfall (mm)'] >= value_75]
```

```python
lst = []

for i in ls_t_r['Rainfall (mm)']:
    
    if i < value_25:
        lst.append("Low")
    
    elif (i >= value_25) & (i <= value_75):
        lst.append("Normal")
    
    elif i > value_75:
        lst.append("High")
        
order = ["High","Normal","Low"]
ls_t_r["Rain_Level"] = lst
ls_t_r["Rain_Level"] = ls_t_r["Rain_Level"].astype('category', categories = order, ordered = True)
```

```python
fig = plt.figure(figsize = (15,10))
sns.set(style="darkgrid")

ax = sns.scatterplot(x = ls_t_r.index,
                     y = ls_t_r["Beef"],
                     hue = "Rain_Level",
                     data = ls_t_r, style = "Rain_Level",
                     size = "Rain_Level",
                     sizes = (50,300),
                     palette = "autumn")

ax = sns.lineplot(x = ls_t_r.index,
                     y = ls_t_r["Beef"].values.tolist(), color = "black"
                     )


ax.set_ylim(0,120000)
plt.xticks(ls_t_r.index.values.tolist())
L = plt.legend(loc = 2,framealpha=2, frameon=True )
L.get_texts()[0].set_text('Rain Level (mm)')
L.get_texts()[1].set_text('High, values above third quartile.')
L.get_texts()[2].set_text('Normal, values between third and first quartile.')
L.get_texts()[3].set_text('Low, values below first quartile.')
ax.set_xlim(1990,2016)
ax.set_xlabel("Years", size = 14)
ax.set_ylabel("Amount of beef produced (Metric Tons)", size = 14)
ax.set_title("Variation of beef produced (Metric Tons) between 1991 and 2015", size = 18)
```
